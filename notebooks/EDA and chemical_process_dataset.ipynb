```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('chemical_process_yield_extended_dataset.csv')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(dataset.corr(),annot=True , cmap='coolwarm')
```

```python
# Distribution of each numerical feature
dataset.hist(figsize=(15, 12), bins=30, color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=dataset, orient='h', palette='Set2')
plt.title("Boxplot of Features")
plt.show()
```

```python
# Scatter plot to identify potential patterns between two main features (customize according to your dataset)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=dataset.columns[0], y=dataset.columns[1], data=dataset)
plt.title("Scatter Plot of Feature 1 vs Feature 2")
plt.show()
```
